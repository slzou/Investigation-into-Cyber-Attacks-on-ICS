```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn import tree
#import os
import time
```

```python
f=open("/home/sharding/Documents/SummerProject/Data/SWaT_Dataset_Normal_v1.csv","r+")
file=csv.reader(f)
top=next(file)
g=open("/home/sharding/Documents/SummerProject/Data/SWaT_Attack.csv","r+")
file_atck=csv.reader(g)
```

```python
colNames=[x.strip() for x in next(file)]
colNames_atck=[x.strip() for x in next(file_atck)]
dataAtck=pd.DataFrame(file_atck)
dataAtck.columns=colNames_atck
dataNorm=pd.DataFrame(file)
dataNorm.columns=colNames
```

# New Data Transformations

```python
actuator_list=["MV101","P101","P102","MV201","P201","P202","P203","P204","P205","P206","MV301","MV302","MV303","MV304","P301","P302","P401","P402","P403","P404","P501","P502","P601","P602","P603"]
```

```python
#Drop constant columns, save to new dataframe
dataNorm_reg_adpt=dataNorm.drop(["Timestamp","Normal/Attack"], axis=1).copy()
dataAtck_reg_adpt=dataAtck.drop(["Timestamp","Normal/Attack"], axis=1).copy()

#Numerise all columns
for x in dataNorm_reg_adpt.columns:
    dataNorm_reg_adpt[x]=pd.to_numeric(dataNorm_reg_adpt[x])
for x in dataAtck_reg_adpt.columns:
    dataAtck_reg_adpt[x]=pd.to_numeric(dataAtck_reg_adpt[x])
    

for x in actuator_list:
    if "P" in x:
        dataNorm_reg_adpt[x]=dataNorm_reg_adpt[x]-1
        dataAtck_reg_adpt[x]=dataAtck_reg_adpt[x]-1
    #else:
        #dataNorm_reg_adpt["{0}_0".format(x)]=[int(v) for v in [u==0 for u in list(dataNorm_reg_adpt[x])]]
        #dataNorm_reg_adpt["{0}_1".format(x)]=[int(v) for v in [u==1 for u in list(dataNorm_reg_adpt[x])]]
        #dataNorm_reg_adpt["{0}_2".format(x)]=[int(v) for v in [u==2 for u in list(dataNorm_reg_adpt[x])]]
        #dataNorm_reg_adpt=dataNorm_reg_adpt.drop([x],axis=1).copy()

for x in dataNorm_reg_adpt.columns:
    if x not in actuator_list:
        dataNorm_reg_adpt[x]=np.hstack((np.array(dataNorm_reg_adpt[x])[1:]-np.array(dataNorm_reg_adpt[x])[:-1],0))
for x in dataAtck_reg_adpt.columns:
    if x not in actuator_list:
        dataAtck_reg_adpt[x]=np.hstack((np.array(dataAtck_reg_adpt[x])[1:]-np.array(dataAtck_reg_adpt[x])[:-1],0))

dataNorm_reg_adpt=dataNorm_reg_adpt[:-1].copy()
dataAtck_reg_adpt=dataAtck_reg_adpt[:-1].copy()
```

```python
#Remove last row for regression, save into new dataframe for forest
AY_rndfor=dataAtck_reg_adpt.copy()
AY_rndformatrix=AY_rndfor.values
#Record columns which we are keeping
AY_cols=AY_rndfor.columns

#Save a separate dataframe for the linear regression, including a linear column
AY_linreg=AY_rndfor.copy()
AY_linreg["Linear"]=1
AY_linregmatrix=AY_linreg.values

#Remove last row for regression, save into new dataframe for forest
Y_rndfor=dataNorm_reg_adpt.copy()
Y_rndformatrix=Y_rndfor.values
#Record columns which we are keeping
Y_cols=Y_rndfor.columns

#Save a separate dataframe for the linear regression, including a linear column
Y_linreg=Y_rndfor.copy()
Y_linreg["Linear"]=1
Y_linregmatrix=Y_linreg.values
```

```python
d_lin={}
for col in actuator_list:
    X = [float(x) for x in list(Y_linreg[col])]
    X_np = np.array(X)
#   X_np = X_np[40:10000]
    Y_lin=Y_linreg.drop([col],axis=1).copy().values
    d_lin["classifier_{0}".format(col)]=linear_model.LinearRegression()
    d_lin["classifier_{0}".format(col)].fit(Y_lin, X_np)
```

```python
normalIndex=list(dataAtck.iloc[list(dataAtck["Normal/Attack"]=="Normal")].index)
attackIndex=list(dataAtck.iloc[list(dataAtck["Normal/Attack"]!="Normal")].index)
```

```python
difference_list={}
for col in actuator_list:
    prediction_matrix=Y_linreg.drop([col],axis=1).values
    dupa=[]
    for x in prediction_matrix:
        m=d_lin["classifier_{0}".format(col)].predict(np.array(x).reshape(1,-1))
        dupa.append(float(m))
    dupa=np.array(dupa)
    difference=dupa-np.array(Y_linreg[col])
    difference_list["diff_{0}".format(col)]=difference
    #plt.figure(figsize=(20,10))
    #for m in attackIndex:
    #    plt.axvline(x=m,color="r",linestyle="-")
    #plt.plot([float(x) for x in list(difference)],".b")
    #plt.title("%s" % col)
    #plt.show()
```

```python
summary_dataframe=pd.DataFrame([],columns=["Part","Mean","Std Deviation","Max"])
for a in actuator_list:
    summary_dataframe.loc[len(summary_dataframe)]=[a,np.mean(difference_list["diff_{0}".format(a)]),np.std(difference_list["diff_{0}".format(a)]),np.max(np.absolute(difference_list["diff_{0}".format(a)]))]
```

```python
for i in range(0,len(summary_dataframe)):
    if summary_dataframe["Std Deviation"][i]==0:
        summary_dataframe["Std Deviation"][i]=0.000001
```

```python
attack_difference_list={}
for col in actuator_list:
    prediction_matrix=AY_linreg.drop([col],axis=1).values
    dupa=[]
    for x in prediction_matrix:
        m=d_lin["classifier_{0}".format(col)].predict(np.array(x).reshape(1,-1))
        dupa.append(float(m))
    dupa=np.array(dupa)
    difference=dupa-np.array(AY_linreg[col])
    attack_difference_list["diff_{0}".format(col)]=difference
```

```python
v=4
```

```python
attack_difference_data=pd.DataFrame([],columns=actuator_list).copy()
for col in attack_difference_data:
    attack_difference_data[col]=attack_difference_list["diff_{0}".format(col)]/float(summary_dataframe.loc[summary_dataframe["Part"]==col]["Std Deviation"])
```

```python
attack_significance_data=pd.DataFrame([],columns=actuator_list).copy()
for col in attack_significance_data:
    a=abs(attack_difference_list["diff_{0}".format(col)]/float(summary_dataframe.loc[summary_dataframe["Part"]==col]["Std Deviation"]))>v
    attack_significance_data[col]=np.multiply(a,abs(attack_difference_list["diff_{0}".format(col)]/float(summary_dataframe.loc[summary_dataframe["Part"]==col]["Std Deviation"])))
attack_significance_data=attack_significance_data.drop(["P201","P202","P401","P404","P502","P601","P603","P101","P403","P602"],axis=1).copy()
```

```python
base=time.strptime('12/28/2015 10:00:00', '%m/%d/%Y %H:%M:%S')
attack_start_times=[]
for t in range(0,len(attack_list)):
    attack_start_times.append(int(time.mktime(time.strptime(attack_list.iloc[t][1], '%m/%d/%Y %H:%M:%S'))-time.mktime(base)))
```

```python
def time_to_detect(t):
    for T in range(t,t+1000):
        if sum(np.array(attack_significance_data.iloc[T])!=0)>=5:
            return(T-t)
    return("No Error Detected")
```

```python
detection_times=[]
for k in attack_start_times:
    detection_times.append(time_to_detect(k))
```

# Testing Models

#### Prediction function

```python
def forest_predict(y,numloops):
    result=pd.DataFrame(columns=Y_cols)
    result.loc[0]=y
    vector=y
    for i in range(1,numloops+1):
        new_vector=[]
        for x in Y_cols:
            new_vector.append(float(d["classifier_{0}".format(x)].predict(vector.reshape(1,-1))))
        new_vector=np.array(new_vector)
        result.loc[i]=new_vector
        vector=new_vector
    return(result)
```

```python
predict_test=forest_predict(Y_rndformatrix[110000],50)
dataNorm_predict=dataNorm_reg.loc[110000:110051].copy()
```

```python
def forest_predict_adv(n,numloops):
    result=pd.DataFrame(columns=Y_cols)
    result.loc[0]=Y_rndformatrix[n]
    vector=Y_adv[n-39]
    for i in range(1,numloops+1):
        new_vector=[]
        for x in Y_cols:
            new_vector.append(float(d_adv["classifier_{0}".format(x)].predict(vector.reshape(1,-1))))
        new_vector=np.array(new_vector)
        result.loc[i]=new_vector
        vector=np.concatenate((vector,new_vector))
        vector=vector[45:]
    return(result)
```

## Changing model

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn import tree
```

```python
f=open("/home/sharding/Documents/SummerProject/Data/SWaT_Dataset_Normal_v1.csv","r+")
file=csv.reader(f)
top=next(file)
g=open("/home/sharding/Documents/SummerProject/Data/SWaT_Attack.csv","r+")
file_atck=csv.reader(g)
colNames=[x.strip() for x in next(file)]
colNames_atck=[x.strip() for x in next(file_atck)]
dataAtck=pd.DataFrame(file_atck)
dataAtck.columns=colNames_atck
dataNorm=pd.DataFrame(file)
dataNorm.columns=colNames
```

```python
colNames
```

```python
#Drop constant columns, save to new dataframe
dataNorm_reg=dataNorm.drop(["Timestamp","Normal/Attack"], axis=1).copy()
dataAtck_reg=dataAtck.drop(["Timestamp","Normal/Attack"], axis=1).copy()

#Numerise all columns
for x in dataNorm_reg.columns:
    dataNorm_reg[x]=pd.to_numeric(dataNorm_reg[x])
for x in dataAtck_reg.columns:
    dataAtck_reg[x]=pd.to_numeric(dataAtck_reg[x])
    
#Remove last row for regression, save into new dataframe for forest
YN_rndfor=dataNorm_reg.copy()
YN_rndformatrix=YN_rndfor.values
YA_rndfor=dataAtck_reg.copy()
YA_rndformatrix=YA_rndfor.values
#Record columns which we are keeping
YN_cols=YN_rndfor.columns
YA_cols=YA_rndfor.columns

#Save a separate dataframe for the linear regression, including a linear column
YN_linreg=YN_rndfor.copy()
YN_linreg["Linear"]=1
YN_linregmatrix=YN_linreg.values
YA_linreg=YA_rndfor.copy()
YA_linreg["Linear"]=1
YA_linregmatrix=YA_linreg.values
```

```python
actuator_list=["MV101","P101","P102","MV201","P201","P202","P203","P204","P205","P206","MV301","MV302","MV303","MV304","P301","P302","P401","P402","P403","P404","P501","P502","P601","P602","P603"]
```

# Makin some regressors

```python
d_lin={}
for col in actuator_list:
    X = [float(x) for x in list(YN_linreg[col])]
    X_np = np.array(X)
#   X_np = X_np[40:10000]
    Y_lin=YN_linreg.drop([col],axis=1).copy().values
    d_lin["regressor_{0}".format(col)]=linear_model.LinearRegression()
    d_lin["regressor_{0}".format(col)].fit(Y_lin, X_np)
```

```python
d_lin["regressor_MV101"].coef_
```

```python
d_for={}
for col in actuator_list:
    X = [float(x) for x in list(YA_rndfor[col])]
    X_np = np.array(X)
#   X_np = X_np[40:10000]
    Y_lin=YA_rndfor.drop([col],axis=1).copy().values
    d_for["regressor_{0}".format(col)]=tree.DecisionTreeRegressor()
    d_for["regressor_{0}".format(col)].fit(Y_lin, X_np)
```

```python
%matplotlib inline
```

```python
difference_list={}
for col in actuator_list:
    prediction_matrix=YA_linreg.drop([col],axis=1).values
    dupa=[]
    for x in prediction_matrix:
        m=d_lin["regressor_{0}".format(col)].predict(np.array(x).reshape(1,-1))
        dupa.append(float(m))
    dupa=np.array(dupa)
    difference=dupa-np.array(YA_linreg[col])
    difference_list["diff_{0}".format(col)]=difference
    plt.figure(figsize=(20,10))
    plt.plot([float(x) for x in list(difference)],".")
    plt.title("%s" % col)
    plt.show()
```

```python
difference_list_rndfor={}
for col in actuator_list:
    prediction_matrix=YA_rndfor.drop([col],axis=1).values
    dupa=[]
    for x in prediction_matrix:
        m=d_for["regressor_{0}".format(col)].predict(np.array(x).reshape(1,-1))
        dupa.append(float(m))
    dupa=np.array(dupa)
    difference=dupa-np.array(YA_rndfor[col])
    difference_list_rndfor["diff_{0}".format(col)]=difference
    plt.figure(figsize=(20,10))
    plt.plot([float(x) for x in list(difference)],".")
    plt.title("%s" % col)
    plt.show()
```

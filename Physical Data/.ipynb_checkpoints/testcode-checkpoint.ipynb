# Importing Data

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f=open("/home/sharding/Documents/SummerProject/Data/SWaT_Dataset_Normal_v1.csv","r+")
file=csv.reader(f)
top=next(file)
g=open("/home/sharding/Documents/SummerProject/Data/SWaT_Attack.csv","r+")
file_atck=csv.reader(g)
```

```python
colNames=[x.strip() for x in next(file)]
colNames_atck=[x.strip() for x in next(file_atck)]
dataAtck=pd.DataFrame(file_atck)
dataAtck.columns=colNames_atck
dataNorm=pd.DataFrame(file)
dataNorm.columns=colNames
```

```python
dataNorm.shape
```

# Transforming Data

```python
%matplotlib inline
```

```python
plt.figure(figsize=(10,5))
plt.plot([float(x) for x in list(dataNorm.PIT501)])
plt.plot([float(x) for x in list(dataAtck.PIT501)])
plt.title("PIT501")
plt.show()
```

```python
for i,j in zip(dataNorm.iteritems(),dataAtck.iteritems()):
    if i[0]!="Timestamp" and i[0]!="Normal/Attack" and j[0]!="Timestamp" and j[0]!= "Normal/Attack":
        plt.figure(figsize=(10,5))
        plt.plot([float(x) for x in list(i[1])])
        plt.plot([float(x) for x in list(j[1])])
        plt.title("%s" % i[0])
        plt.show()
```
